```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# Parameters
N = 30  # number of Fock states
r = 1.0  # squeezing parameter
xvec = np.linspace(-5, 5, 200)

# Create the vacuum state
vacuum_state = basis(N, 0)

# Apply the squeezing operator to the vacuum state
squeezed_vacuum_state = squeeze(N, r) * vacuum_state

# Calculate the Wigner function
W_squeezed_vacuum = wigner(squeezed_vacuum_state, xvec, xvec)

# Plot the Wigner function
fig, ax = plt.subplots(figsize=(8, 6))  # Create a single plot with an 8x6 inch figure size
cont = ax.contourf(xvec, xvec, W_squeezed_vacuum, 100, cmap='Spectral')

# Adjusting the font size of labels
ax.set_xlabel('x', fontsize=14)  # Label the x-axis with increased font size
ax.set_ylabel('p', fontsize=14)  # Label the y-axis with increased font size

# Increase font size of colorbar labels
cbar = fig.colorbar(cont, ax=ax, orientation='vertical')
cbar.ax.tick_params(labelsize=14)  # Increase font size of the colorbar ticks

plt.tight_layout()  # Adjust the layout to make it look neat
plt.show()  # Display the plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# Parameters
N = 15  # number of Fock states (lower to reduce computation time for two-mode system)
r = 1.0  # squeezing parameter
xvec = np.linspace(-5, 5, 200)  # x-axis for phase space

# Create the two-mode vacuum state (tensor product of two vacuum states)
vacuum_state = tensor(basis(N, 0), basis(N, 0))

# Create the two-mode squeezing operator
S = squeeze(N, r)  # single mode squeezing operator
S_two_mode = tensor(S, S)  # applying squeezing to both modes

# Apply the two-mode squeezing operator to the vacuum state
two_mode_squeezed_state = S_two_mode * vacuum_state

# Calculate the Wigner function for both modes together
W = wigner(two_mode_squeezed_state, xvec, xvec)

# Plot the Wigner function for the two-mode squeezed state
fig, ax = plt.subplots(figsize=(8, 6))
cont = ax.contourf(xvec, xvec, W, 100, cmap='Spectral')

# Adjusting the font size of labels
ax.set_xlabel('x', fontsize=14)  # Label the x-axis with increased font size
ax.set_ylabel('p', fontsize=14)  # Label the y-axis with increased font size

# Increase font size of colorbar labels
cbar = fig.colorbar(cont, ax=ax, orientation='vertical')
cbar.ax.tick_params(labelsize=14)  # Increase font size of the colorbar ticks

plt.tight_layout()  # Adjust the layout to make it look neat
plt.show()  # Display the plot
```
